```python
import numpy as np
```

```python
matrix = np.random.randint(9, size=(7, 7))
matrix
```

```python
ar = np.empty((7, 7))
n = matrix.shape[0]
m = matrix.shape[1]
```

```python
for i in range(n):
    for j in range(m):
        s = 0.0
        count = 0
        if i > 0:
            s += matrix[i - 1, j]
            count += 1
        if i < n - 1:
            s += matrix[i + 1, j]
            count +=1
        if j > 0:
            s += matrix[i, j - 1]
            count += 1
        if j < m - 1:
            s += matrix[i, j + 1]
            count +=1
        ar[i,j] = s / count    
            
        
```

```python
ar = np.around(ar, 1)
ar
```

```python
n, m = ar.shape
s = 0.0
for i in range(n):
    for j in range(m):
        if i > j:
            s += ar[i, j]
print(s) 
```
